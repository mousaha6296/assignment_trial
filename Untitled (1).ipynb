```python
1. What is the difference between a function and a method in Python?
A function is a standalone block of reusable code that performs a specific task. For example, a function called add might take two numbers as input and return their sum.

A method, on the other hand, is associated with an object. It is called on that object and can manipulate or retrieve data from it. For example, a method like upper() is called on a string to convert all its characters to uppercase.

2. Explain the concept of function arguments and parameters in Python.
Parameters are placeholders defined in the function’s declaration. For instance, if a function is defined as add_numbers(a, b), a and b are parameters.

Arguments are the actual values you pass to the function when calling it. For example, if you call add_numbers(2, 3), 2 and 3 are the arguments.

3. What are the different ways to define and call a function in Python?
A function can be defined normally using the def keyword, such as a function called square that calculates the square of a number. To call this function, you would pass a number as an argument, like square(4) to get 16.

Alternatively, you can define short, anonymous functions using the lambda keyword. For example, a lambda function to calculate the square can also be defined as lambda x: x * x. You can then call it the same way, by passing a number like 4.

4. What is the purpose of the return statement in a Python function?
The return statement sends the result of a function back to the caller. For example, if a function calculates the product of two numbers, the return statement ensures that the result is available for use outside the function. Without it, the function would not provide any output to the caller.

5. What are iterators in Python and how do they differ from iterables?
An iterable is something that can be looped over, like a list of numbers or a string of characters. For example, you can loop through the numbers in a list one by one.

An iterator is a tool used to perform this looping. It produces the values one at a time when requested, making it efficient for handling large datasets. For instance, if you have a list, you can convert it into an iterator, and each time you ask for the next value, the iterator will provide it.

6. Explain the concept of generators in Python and how they are defined.
A generator is a type of iterator that generates values on the fly, rather than storing them all in memory at once. This makes generators useful when working with large datasets or infinite sequences. For example, a generator could produce a sequence of numbers from 1 to 100 without needing to store all those numbers in memory.

7. What are the advantages of using generators over regular functions?
Generators are more memory-efficient because they don’t store all values in memory. Instead, they calculate each value when needed. For example, if you use a generator to read a large file line by line, it will only keep the current line in memory, instead of loading the entire file.

8. What is a lambda function in Python and when is it typically used?
A lambda function is a compact way to define a function for a short task. For example, if you need to double a number, you can define a lambda function instead of a full function. Lambda functions are often used in data analytics for operations like sorting or filtering data based on certain conditions.

9. Explain the purpose and usage of the map() function in Python.
The map() function is used to apply a given function to every item in a list or other iterable. For example, if you have a list of numbers, you can use map() to square each number in the list, and it will return the list of squared values.

10. What is the difference between map(), reduce(), and filter() functions in Python?
map() is used to apply a function to all elements in a dataset. For example, you can double all numbers in a list using map().
reduce() combines all elements into a single value, such as calculating the product of all numbers in a list.
filter() is used to select elements that satisfy a condition, such as picking only even numbers from a list.

11.
```

```python
# Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.
```

```python
def sum_of_evens(numbers):
    return sum(num for num in numbers if num % 2 == 0)
```

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

primes = [n for n in range(1, 201) if is_prime(n)]
```

```python
#Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms
```

```python
class FibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.n_terms:
            value = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return value
        else:
            raise StopIteration

# Example usage:
fib = FibonacciIterator(10)
print(list(fib))
```

```python
# Write a generator function in Python that yields the powers of 2 up to a given exponent.
```

```python
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i
```

```python
# Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()

# Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
tuples = [(1, 3), (4, 1), (2, 2)]
sorted_tuples = sorted(tuples, key=lambda x: x[1])


# Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
def celsius_to_fahrenheit(celsius):
    return list(map(lambda c: (c * 9/5) + 32, celsius))


# Create a Python program that uses `filter()` to remove all the vowels from a given string
def remove_vowels(string):
    return ''.join(filter(lambda x: x.lower() not in 'aeiou', string))
```
